#### Data Visualization With Seaborn
Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code. In this lesson, we will cover the basics of Seaborn, including creating various types of plots and customizing them.

````python
import seaborn as sns
````

````python
### Basic Plotting With Seaborn
tips=sns.load_dataset('tips')
tips
````

````python
##create a scatter plot
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
````

````python
## Line Plot

sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line Plot of Total bill by size")
plt.show()
````

````python
## Categorical Plots
## BAr Plot
sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar Plot of Total Bill By Day')
plt.show()
````


### When using sns.barplot from the Seaborn library, the function calculates the mean value of the numerical variable (total_bill in this case) for each category of the categorical variable (day), and then plots these averages as bars.

````python
## Box Plot
sns.boxplot(x="day",y='total_bill',data=tips)
````

A boxplot (also known as a box-and-whisker plot) is a statistical graphic that visually represents the distribution of a dataset. When you use `sns.boxplot` in Seaborn, it helps you understand the spread, central tendency, and potential outliers for the `total_bill` values grouped by `day`. 

Hereâ€™s how to interpret the boxplot:

---

### **Components of the Boxplot**
1. **Box:**
   - **Lower Edge (Q1):** Represents the first quartile (25th percentile). 25% of the data falls below this value.
   - **Upper Edge (Q3):** Represents the third quartile (75th percentile). 75% of the data falls below this value.
   - **Line in the Box (Median):** Represents the median (50th percentile). Half the data lies below and half above this value.

2. **Whiskers:**
   - Extend from the box to show the range of data within 1.5 times the interquartile range (IQR).  
   - **IQR = Q3 - Q1**.
   - The whiskers represent the smallest and largest non-outlier data points within this range.

3. **Outliers (Points outside the whiskers):**
   - Data points that fall beyond 1.5 Ã— IQR from the Q1 or Q3 are considered outliers and are plotted as individual points.

---

### **How the Boxplot for `sns.boxplot(x="day", y="total_bill", data=tips)` Works:**
- **`x="day"`**: The categorical variable (`day`) is plotted on the x-axis.
- **`y="total_bill"`**: The numerical variable (`total_bill`) is plotted on the y-axis.
- For each unique value of `day` (e.g., "Sun", "Sat", "Thur"), the boxplot summarizes the `total_bill` data:
  - The box shows the IQR of the total bills for that day.
  - The line inside the box represents the median total bill.
  - Whiskers show the typical range (excluding outliers).
  - Outliers, if any, are plotted as individual points.

---

### **Usefulness of the Boxplot**
- Identifying the spread of data for each category (`day` in this case).
- Comparing the medians and IQRs across categories.
- Detecting skewness or asymmetry in the data distribution.
- Highlighting outliers that deviate significantly from the rest of the data.

````python
## Violin Plot

sns.violinplot(x='day',y='total_bill',data=tips)
````

A **violin plot** is a combination of a **boxplot** and a **kernel density plot**, providing a richer visualization of the data distribution. When you use `sns.violinplot(x='day', y='total_bill', data=tips)`, the plot shows how the distribution of `total_bill` varies across different days (`day`).

---

### **Components of a Violin Plot**
1. **Kernel Density Estimation (KDE):**
   - The "violin" shape represents a smoothed version of the data distribution for each category on the x-axis (`day` in this case).
   - The width of the violin indicates the density of the data at different `total_bill` values. 
   - A wider section means more data points in that range, and a narrower section means fewer data points.

2. **Inner Boxplot:**
   - Inside each violin, there is often a **boxplot** (or similar markers) that shows summary statistics:
     - **White dot**: Represents the median of the data.
     - **Thick bar**: Represents the interquartile range (IQR).
     - **Thin line**: Represents the whiskers (data within 1.5 Ã— IQR).

3. **Symmetry:**
   - By default, the violin is symmetric about its vertical axis. However, you can modify it to display only one side (e.g., `split=True` for paired data).

---

### **How This Works for `sns.violinplot(x='day', y='total_bill', data=tips)`**
- **x='day'**: Days (`Sun`, `Sat`, `Thur`, etc.) are plotted on the x-axis as categories.
- **y='total_bill'**: `total_bill` values are plotted on the y-axis as the numerical variable.
- For each day:
  - The violin shape shows the distribution of `total_bill` values for that day.
  - The inner boxplot shows key statistics (median, quartiles, whiskers).

---

### **Advantages of Violin Plot**
- Provides a detailed view of the data distribution, unlike a boxplot, which only shows summary statistics.
- Highlights multimodal distributions (e.g., data with multiple peaks).
- Combines the benefits of boxplots and KDEs, offering both summary statistics and a sense of data density.

---

### **Customizations**
- **Add hue for subcategories:**
  ```python
  sns.violinplot(x='day', y='total_bill', hue='sex', data=tips, split=True)
  ```
  This splits the violins by `sex` and shows the distribution for males and females separately.
  
- **Control inner representation:**
  ```python
  sns.violinplot(x='day', y='total_bill', data=tips, inner='stick')
  ```
  Changes the inner markings to show individual data points instead of a boxplot.

````python
import pandas as pd
sales_df=pd.read_csv('sales_data.csv')
sales_df.head()
````

````python
## Plot total sales by product
plt.figure(figsize=(10,6))
sns.barplot(x='Product Category',y="Total Revenue",data=sales_df,estimator=sum)
plt.title('Total Sales by Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show()
````

````python
## Plot total sales by Region
plt.figure(figsize=(10,6))
sns.barplot(x='Region',y="Total Revenue",data=sales_df,estimator=sum)
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
````


### Exactly! The estimator parameter in sns.barplot controls the aggregation function applied to the y values for each unique category in the x axis. By default, sns.barplot uses the mean as the estimator. However, when you set estimator=sum, it computes the sum of the y values instead.
